```python
# Importing the required libraries

from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
page = requests.get("https://en.wikipedia.org/wiki/Main_Page")
page
```

```python
page.content
```

```python
soup = BeautifulSoup(page.content)
soup
```

```python
print(soup.prettify())
```

```python
heading = soup.find_all(["h1","h2","h3","h4","h5","h6"])
heading
```

```python
new_heading=[]

for i in heading:
    new_heading.append(i.text.replace("\n", " "))
new_heading
```

```python
# Top 100 Imdb Movies

movies = requests.get("https://www.imdb.com/search/title/?groups=top_100&sort=user_rating,desc")
movies
```

```python
soup = BeautifulSoup(movies.content)
soup
```

```python
# Top 100 movies
name = soup.find_all("h3", class_="lister-item-header")
name
```

```python
movies_name = []

for i in name :
        movies_name.append(i.text.replace("\n",""))
        
movies_name[:100]
```

```python
# Top 100 Imdb Indian Movies

movies = requests.get("https://www.imdb.com/india/top-rated-indian-movies/")
movies
```

```python
soup = BeautifulSoup(movies.content)
soup
```

```python
movies = soup.find_all('td',class_="titleColumn")
movies
```

```python
movies_name = []

for i in movies :
        movies_name.append(i.text.replace("\n", " "))
        
movies_name[:100]
```
